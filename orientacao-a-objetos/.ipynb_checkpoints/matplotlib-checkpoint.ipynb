# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50) # vai gerar números inteiros de 1000 a 3000, 50 vezes
meses = np.arange(1, 51) # vai gerar um range de números que vai do 1 até 50
```

- Editando o Gráfico de Linha

```python
plt.plot(meses, vendas)
plt.axis([0, 50, 0, max(vendas)+200])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?

# eu copiei o código acima e apenas acrescentei '-' que representa esse tipo de gráfico e podemos adicionar cores
plt.axis([0, 50, 0, max(vendas)+200])
plt.plot(meses, vendas, '-', color='pink') 
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Dispersão

```python
# em vez de colocar 'plot', coloque 'scatter' que vira um gráfico de dispersão
plt.scatter(meses, vendas, color='purple')
plt.show()
```

- Barras

```python
# em vez de colocar 'plot', coloque 'bar' que vira um gráfico de barras
plt.bar(meses, vendas, color='yellow')
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
# figure = é uma figura que vai ser colocado os gráficos
# figsize = largura/altura
plt.figure(1, figsize=(17, 5))

# os números = quantas linhas/quantas colunas/posição a ser preenchida
plt.subplot(131)

# primeiro subplot
plt.plot(meses, vendas, '-', color='pink') 

# segundo subplot
plt.subplot(132)
plt.scatter(meses, vendas, color='purple')

# terceiro subplot
plt.subplot(133)
plt.bar(meses, vendas, color='yellow')

# mostrando os gráficos
plt.show()
```
